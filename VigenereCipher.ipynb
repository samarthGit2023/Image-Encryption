```python
from PIL import Image

# Vigenère cipher encryption function
def vigenere_encrypt(plaintext, key):
    ciphertext = bytearray()
    key_length = len(key)
    for i, byte in enumerate(plaintext):
        key_byte = ord(key[i % key_length])
        encrypted_byte = (byte + key_byte) % 256
        ciphertext.append(encrypted_byte)
    return bytes(ciphertext)

# Open the image file
image_path = "madara.jpg"  # Replace with the actual path to your image
image = Image.open(image_path)

# Get the bytes of the image
image_bytes = image.tobytes()

# Encryption key
key = "aaaaaaaazzzzzzzzxxxxxxxttttttttbbbbbbbboooooooopppppppp"

# Encrypt the image bytes
encrypted_bytes = vigenere_encrypt(image_bytes, key)

# Create a new image from the encrypted bytes
encrypted_image = Image.frombytes(image.mode, image.size, encrypted_bytes)

# Save the encrypted image
encrypted_image_path = "path_to_save_encrypted_image.png"  # Replace with the desired path and filename
encrypted_image.save(encrypted_image_path)

print("Encrypted image saved successfully.")
```



*   In the above code, we use **Vigenère cipher** for the encryption of the input image.


*   In this encryption algorithm, we convert image to a series of bytes of the grayscale data or RGB data. We perform operations on these bytes to obtain a series of bytes of encrypted image.

* We choose any string as our key and pass it to the vigenere_encrypt function along with the array of bytes of image. Each byte of image_bytes represents grayscale value or RGB value of the corresponding pixel which varies from 0 to 255.

* In the function we convert each symbol of key into corresponding value using **ord()** and use that value to encrypt the pixel byte data by performing the following operation :
* **encrypted_byte = (byte + key_byte) % 256**
* Here key_byte is the value of a symbol of key, byte is the value of pixel in range (0, 255). We add the key_byte to value of pixel and take mod with 256 to attain encrypted value of pixel.
